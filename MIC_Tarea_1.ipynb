Ahora, teniendo un poco de entendimiento de la función seno, con el siguente código se puede graficar la función entre $0$ y $2\pi$.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6), dpi=80)
x=np.arange(0.0,2*np.pi,0.5)
y=np.sin(x)
plt.plot(x,y)
plt.grid(axis='both')
plt.xticks([0, np.pi/2, np.pi, 3/2*np.pi, 2*np.pi],
          [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])
plt.yticks([-1,0,1],
          [r'$-1$', r'$0$', r'$1$'])
plt.show()
```
